## C# Dili Genel Özellikleri

C# (Si Şarp diye okunur) programlama dili 2002 yılında Microsoft tarafından C# 1.0 versiyonu ile kullanıma
sunulmuş çok amaçlı bir programlama dilidir. C# ile masaüstü yazılımlar, konsol uygulamaları, web
uygulamaları, web servisleri, mobil uygulamalar, oyunlar ve makine öğrenmesi modelleri
geliştirilebilmektedir (Whitaker, 2022).

C#, C dil ailesine mensuptur. Bu sebeple dilin gramer yapısı/söz dizimi (sentaks) C, C++, Java ve JavaScript
dillerine benzemektedir. Bu dillerden herhangi birine aşina bir programcı kolaylıkla C# diline kısa sürede
uyum sağlayabilir veya C# bilen bir geliştirici bu sayılan dilleri kısa sürede öğrenebilir.
Bir programlama dilinde yazılan kaynak kodların (kısaca kod) bilgisayarda çalışabilmesi için kodun
bilgisayarın anlayacağı dile yani makine diline çevrilmesi gerekir. Kaynak kodun makine diline çevrimi
derleme (compiling) ve yorumlama (interpreting) olarak bilinen iki farklı şekilde olmaktadır. Buradan
hareketle programlama dilleri derlenen veya yorumlanan diller olarak da sınıflandırılabilmektedir (Özen,
2012).

C# derlenen bir dildir. Bunun anlamı C# kodlarının bilgisayar üzerinde çalışabilen bir programa
dönüşebilmesi için derleyici (compiler) adındaki bir yardımcı programın kaynak kodu makine diline veya ara
bir formata dönüştürmesi gerekir. C# dilinde yazılan kodun derlenmesi ve C# diliyle yazılan programların
çalışması için .NET platformunun kurulu olması gerekir.

C# dilinde kaynak kodun derlenmesi iki aşamalıdır. Kaynak kod .NET tarafından önce ara bir forma çevrilir.
Program çalıştırılmak istendiğinde ise ara dildeki kod yine .NET platformunun bir parçası olan CLR
(Common Language Runtime - Ortak Dil Çalışma Platformu) aracılığıyla ikinci bir derleme işlemi sonrası
makine diline çevrilir ve program çalışır (Doyle, 2015).

## Dilin Kod Yapısı

Bir kitabı veya dokümanı yukarıdan aşağıya okuduğumuz gibi kaynak kod dosyasını bilgisayarın anlayacağı
dile çeviren derleyici veya yorumlayıcı dediğimiz yardımcı programlar da kaynak kodu yukarıdan aşağıya
doğru okurlar. Bir kaynak kod dosyası alt alta satırlarda veya aynı satır içinde yan yana birden fazla farklı
işlevde kod kısımları içerir. Derleyicinin her biri farklı bir işlevde olan bu kodları birbirinden ayırması ve iş
adımlarını sırayla işleyebilmesi için noktalı virgül ; kullanılır. ; operatörü iş talimatının bittiğini kendisinden
sonraki ifadelerin yeni bir iş talimatı olduğunu ifade eder.

C# dilinin kod yapısı aşağıdaki şekilde verilen düzeni yukarıdan aşağı takip etmektedir

    * namespace
    * class
    * metod
    * blok
    * deyim
    * ifade
     c# kod mimarisi

C# nesne tabanlı bir dil olduğundan olay örgüsü sınıflar içerisinde gerçekleşir ve her bir sınıf (class) bir alan
uzayı (namespace) içinde yer almaktadır. Örneğin konsol uygulamasında kullanılan Console sınıfı System
alan uzayı içindedir bu sebeple Console sınıfını kaynak kodda kullanabilmek için System alan uzayı using
System; kullanımıyla koda dahil edilmelidir. Sınıflar bir veya daha fazla farklı işlevleri olan metodlardan
oluşur.
MerhabaDunya isimli ilk programın kaynak kodları üzerinden dilin kod yapısını inceleyelim.

```c#

            Console.WriteLine("Merhaba Dünya!");}
            }
            }
```

Koda bakıldığında bazı satırların diğerlerine göre daha içeriden yazıldığı fark edilecektir. Bu girinti, kodun
daha iyi anlaşılması ve okunurluğunu artırmak içindir. Tüm satırlar aynı hizada yazıldığında da hiçbir
problem olmaz, program yine sorunsuz derlenir. İlave olarak programın birden fazla satırdan oluştuğu
görülecektir. Satırlar da aynı şekilde kodun okunurluğunu artırmak, farklı yapıdaki kod birimlerini ve
bağımsız işleri birbirinden ayırmak içindir. Tüm satırlardaki boşluklar ve yeni satırlar kaldırılıp tek bir satır
haline getirildiğinde program yine sorunsuz derlenecek ve çalışacaktır. Aşağıdaki resimde program tek satır
haline getirilmiş ve satırlardaki girinti boşlukları kaldırılmıştır. F5 tuşuna veya içi yeşil üçgen butona
basıldığında kaynak kod sorunsuz derlenecek ve program çalışacaktır.

En içerideki kod satırından başlayarak kodları ve C# kod yapısını anlamaya çalışalım. En içerideki
Console.WriteLine("Merhaba Dünya!"); satırındaki Console.WriteLine() metodu konsol ekranına bir şey
yazmak yani çıktılamak için kullanılır. Console.WriteLine() metodu parantezler içinde verilen metni ekrana
basar.

Satırın sonunda noktalı virgül ; olduğu görülecektir. Algoritma geliştirirken birbirinden bağımsız her bir işi
ayrı bir adımda veriyorduk. Böylece problemin çözümüne yönelik işleri birbirinden ayırmakta ve uygun
şekilde sıralamaktaydık. Programlama süreci de algoritmik akışı bir programlama dilinde ifade etmektir.
Programlarken de benzer şekilde her bir iş adımını bir diğerinden ayırmalıyız. İşte noktalı virgül ; C# dilinde
iş adımlarını birbirinden ayırmak için kullanılan operatördür. Böylece derleyici kodda kaç iş olduğunu ;
operatörüne bakarak algılar ve kaynak kodu yukarıdan aşağıya doğru sırayla çalıştırır.

Bu satırın üstünde bir süslü parantez ve onun üstünde static void Main(string[] args) satırının olduğu
görülecektir. Main(), program açıldığında ilk çalıştırılacak metodun adıdır. Web sitelerindeki ana sayfanın
benzeridir. Tarayıcıya bir web sitesinin adını yazdığımızda bizi o web sitesinin ana sayfası karşılar. Main()
metodu da C# dilinde yazılan programlarda aynı işlevi görür. Bir C# programı çalışmaya öncelikle Main()
metodunu çağırarak yani Main() metodundaki kodları yürüterek başlar.

Main() metodunun üzerinde süslü parantez ve onun üzerinde internal class Program satırı yer almaktadır. C#
dili nesne tabanlı bir dildir; kod tasarımı ve tüm akış sınıf kavramı üzerinden cereyan eder. Her C#
programında en az bir adet sınıf olmalıdır. Bu satırda da sınıfımızın adı Program olarak verilmiştir.
internal class Program satırının üzerinde yine süslü parantez ve onun üzerinde namespace MerhabaDunya
satırı yer almaktadır. Bir projede birden fazla sınıf olabilir. Bu sınıfları hiyerarşik düzende ve mantıksal
olarak bir arada tutmak için namespace (alan uzayı) adı verilen yapı kullanılır. Bu satırda programdaki sınıfı
kapsayan alan uzayının adı tanımlanmaktadır.

Şimdi de programı bu verilen bilgiler ışığında yukarıdan aşağıya doğru okuyalım.

Her C# programı en az bir sınıf (class) içermelidir ve sınıflar alan uzayı adı verilen yapı içinde (namespace)
tanımlanır. Projeye ad olarak verdiğimiz MerhabaDunya tüm her şeyi torbalayan alan uzayımızın da ismidir.
Alan uzayı en az bir tane olmak kaydıyla sınıfları barındırmaktaydı ve ilk uygulamamızdaki sınıfımızın adı
Program olarak verilmiştir.

Sınıflar ise her biri ayrı işlevi yerine getiren metodlardan oluşur. C# dilinde
kodlanan program çalıştırıldığında Main adındaki metodu arar ve çalışmaya oradan başlar. Main metodunun
içine bakıldığında ekrana bir şey çıktılayan Console.Writeline() komutunun olduğu görülecektir. Komutun
çıktılayacağı metin ise çift tırnak içinde verilen Merhaba Dünya! dizgesidir.

Programlama yaparken kaynak kod, bloklar halinde yazılmaktadır. Kod blokları süslü parantezler {} içinde
yer alan kod satırlarıdır.

Kod mimarisini açıklarken namespace, class, Main metodu gibi farklı dil yapılarından bahsedildi. İşte süslü
parantezler her bir kod yapısının içindeki satırları sarmalama işlevi görür. Süslü parantezlerin kapsadığı
satırlara kod bloğu denir. Böylece büyükten küçüğe doğru namespace, class, Main metodu, diğer metodlar,
ileride göreceğimiz karar ve döngü blokları süslü parantez içinde yazılır. Süslü parantezler hem kodu
gruplama işlevi görürler hem de programda bir hata olması durumunda programcının hangi yapıya
odaklanacağını işaret ederler.

Bir kod bloğu bir diğer kod bloğunun içerisinde yer alabilir. Buna iç içe blok denir Verilen program
örneğinden de görüleceği üzere yapı olarak daha büyük olan kendisinden küçük olanı kapsamaktadır.

İkinci satırdaki x = x + 1; denklemi matematiksel açıdan geçersiz iken programlama dilinde bir değişkenin
bellekteki değerine bir ekleyip yine o değişkene yeni toplam değeri atamaktan ibarettir. Bu yapı
programlamada sıklıkla kullanılmaktadır.

x = x + 1; satırını anlamak için değişken kavramını tekrar hatırlamak gerekmektedir. Değişken, bellekte
programın ihtiyaç duyduğu veriyi saklaması için ayrılmış alanları ifade etmekteydi. Adı üzerinde bu bellek
alanında saklanan değer daha sonradan değiştirilebilir özelliktedir. Bu kısa hatırlatmadan sonra koda göz
attığımızda bir atama operatörü olduğunu görmekteyiz. Atama operatörünün çalışma mekanizması iki
aşamalıdır. Öncelikle atama operatörünün sağ tarafı hesaplanır/değerlendirilir ve ardından ikinci iş olarak
hesaplanan değer atama operatörünün sol tarafındaki değişkene atanır. İlk aşamada eşitliğin sağ tarafı
hesaplanır ve 6 değeri elde edilir. İkinci aşamada ise x değişkenine 6 değeri atanacaktır. Görüldüğü üzere
matematiksel açıdan geçersiz olan bu denklem programlama literatüründe basit bir atama işleminden başkası
değildir.

#### Değişkenlerin Kapsamı

Değişkenlerin uygulama içinde erişilebilir yani etkili olduğu bölge değişkenlerin kapsamı/kapsama alanı
olarak tanımlanır. Tanımlanan herhangi bir değişken uygulamanın her yerinden erişilebilir değildir.
Değişkenlerin kapsamı tanımlandığı yere göre değişmektedir. Değişkenlerin üç farklı kapsamı bulunmaktadır
(Szabó, 2020).

    1.  Sınıf Seviyesi : Bir sınıfın içinde tanımlanan değişkenler o sınıfın herhangi bir yerinden erişilebilir özelliktedir, sınıfın herhangi bir metodu tarafından çağrılabilir, kullanılabilir ve değiştirilebilirler. Bu sınıf değişkenlerine sınıf  üyesi (class member veya field) denilmektedir. Aşağıdaki programda string veri tipinde dersAdi isimli bir sınıf değişkeni tanımlanmıştır. Bu sınıf değişkenine sınıfın metodları erişebilir ve değiştirebilirler.

```c#
    using System;
    namespace SinifSeviyesi
     {
    class Program
    {
        string dersAdi = "C# Programlama"; // sinif seviyesi degisken
         public void Ciktila()
        {
        // sinif seviyesi degiskenin degeri degistiriliyor
        dersAdi = "Görsel Programlama";
         Console.WriteLine(dersAdi);
               }
          }
 }
 
```
   
    2.  Metod Seviyesi : Bir metod içinde tanımlanan değişkenler yalnızca o metod içinde kullanılabilirler. Metod seviyesi değişkenler tanımlandığı metodun dışında başka bir metod içinden veya daha yüksek seviye bir yapı (sınıf,
    alan uzayı) tarafından erişilemez, değiştirilemez. Aşağıdaki programda Ciktila isimli metodda tanımlanan ushort veri tipinde metodLevelSayi isimli değişkene yalnızca tanımlandığı Ciktila isimli metod içinden erişilebilir, Degistir metodu içinde bu değişkene erişilemez veya değeri değiştirilemez.

```c#
    using System;
    namespace MetodSeviyesi
        {
    class Program
        {
        public static void Ciktila()
            {
            ushort metodLevelSayi = 2023; // metod seviyesi degisken
            Console.WriteLine(metodLevelSayi);
                    }
                public void Degistir()
                {
                    // Console.WriteLine(metodLevelSayi); // bu satir hata verecektir
                }
                 static void Main(string[] args)
                {
                Console.WriteLine("Ciktila Metodu cagriliyor...");
                Ciktila();
                Console.ReadLine();
                }
            }
        }
```


    3.  Blok Seviyesi  : Bir blok içinde tanımlanan değişkenler yalnızca tanımlandığı bloktan erişilebilir durumdadır. O bloğu dışında değişkene erişilemez veya değiştirilemez. Döngü yapıları içinde kullanılan indisler blok seviye değişkenleridir. Dış blokta tanımlı bir değişkene iç bloktan erişilebilir fakat iç blokta tanımlı değişkene dış bloktan erişilemez.
#### Erişim Belirleyiciler : 

Bazen program içindeki tip, metod, değişken, struct gibi yapıların son kullanıcı tarafından veya sınıfın bazı
metodları tarafından veya diğer sınıflar tarafından erişilememesini istediğimiz durumlar olabilir. Örneğin
bazı değişkenler sınıf içi yapıyı tanzim etmek için kullanılırlar bu nedenle dışarıdan bu değişkenlerin
değerlerinin görüntülenmesi anlamsız olacaktır. Kimi zaman bazı metodlar gizli bilgi taşırlar ve bu nedenle
sadece izinli metodların bu bilgiye erişiminin olmasını isteriz, diğer metod veya yapıların bu metoda
erişmesini istemeyiz. İşte bu amaçla bazı programlama yapılarına erişimi sınırlamak için erişim
belirleyiciler (access modifiers) adı verilen 4 farklı yapı 6 farklı seviye kullanılmaktadır. Aşağıda listelenen
erişim belirleyiciler tip, değişken, metod ve struct dil yapılarının sınıf içinde, sınıf dışında, proje içinde veya
dışında görünürlüğünü etkilemektedir.

**public**:

public anahtar kelimesi ile tanımlanan bir dil bileşeni her yerden görünür ve erişilebilirdir. Tanımlandığı
sınıftan, diğer sınıflardan, farklı derlemedeki sınıflar tarafından erişilebilir ve kullanılabilir.

**private**:

private anahtar kelimesi ile tanımlanan bir dil bileşeni yalnızca tanımlandığı sınıf veya struct tarafından
erişilebilir. private, değişkenlerin varsayılan erişim belirleyicisidir.

**protected**:

protected anahtar kelimesi ile tanımlanan sınıf, metod veya sınıf değişkenleri yalnızca tanımlandığı sınıf
veya bu sınıftan türetilmiş sınıflardan veya farklı derlemede yer alan türetilmiş sınıflardan erişilebilir.

**internal**:

internal anahtar kelimesi ile tanımlanan bir dil bileşeni hangi kod bileşenleri ile birlikte derlendiyse onlar
tarafından erişilebilir. internal, sınıfların varsayılan erişim belirleyicisidir.

**protected internal**:

protected internal anahtar kelimesi ile tanımlanan bir dil bileşeni aynı derlemede yer alan sınıf veya
türetilmiş sınıflar tarafından veya başka bir derlemede yer alan türetilmiş sınıf tarafından erişilebilir.

**private protected**:

private protected anahtar kelimesi ile tanımlanan bir dil bileşeni derlendiği sınıftan veya aynı derlemede
türetilmiş diğer sınıflar tarafından erişilebilir

### Sabitler (Constant):

Kimi zaman program içinde asla değişmesini istemediğimiz değerler olabilir. Bu gibi durumlarda sabit
(constant) adı verilen programlama bileşeni kullanılır (Wagner, 2021). Sabitlere girilen değerler adı üzerinde
asla değiştirilemez. C# dilinde sabit tanımlamak için const anahtar kelimesi kullanılır. Bir sabit tanımlandığı
esnada ona değer atanmalıdır. Sabitin değeri daha sonra değiştirilmek istendiğinde program hata verir.

```c#
const double isikYili = 9.46E12;
```

**Daha önce tanımlanmış bir değişkenin değeri sabite atanamaz. Aşağıdaki sabit değer ataması hata verecektir**

```c#
int x = 100;
const long hesap = x; // gecersiz
```

### Değerler (Literal):

Program içinde kullanılacak değerlerin bilgisayarda saklanması için değişken adı verilen kullanılan yapının
kullanıldığı ifade edilmişti. Değişkenlere atanan değerler, kullanıcı tarafından girilen değerler veya bir veri
kaynağından çekilip program içinde kullanılan her türlü veri programlama literatüründe değer (literal) olarak
isimlendirilir. Bir veri tipine atanacak değer, o veri tipiyle uyumlu olmalıdır.

C# dilinde veri tiplerinin alabileceği değerler aşağıda kategoriler halinde açıklanmıştır.

    **Nümerik Değerler:**

    * Nümerik veri tiplerine değer girilirken rakamlar kullanılır. 
    * Eğer sayı onaltılık tabanda ise ilave olarak A-F arası harfler, 
    * sayının ondalık ayıracı olarak nokta . operatörü, 
    * bilimsel gösterim (notasyon) için E harfi,
    * sayının okunurluğunu artırmak için istenirse binlik ayıracı olarak alt çizgi _ karakteri kullanılmaktadır.

    Bunların haricinde boşluk, virgül karakterleri nümerik değerlerde kullanılmaz. 
    long, float ve decimal veri tipi
    tercih edilmişse bu veri tiplerine girilen değerin sonuna sırasıyla (küçük veya büyük) L, F ve M harfleri
    eklenmelidir. Sayının işareti için – ve + operatörleri kullanılır.

    Günlük hayatta hesaplamalarımızı genellikle onluk sistem üzerinden yaparız. C# dilinde de tam sayı veri
    tiplerine değer atanırken girilen değerler varsayılan olarak onluk tabanda değerlendirilmektedir. Bununla
    birlikte değişkene atanan sayının farklı tabanlarda değerlendirilmesini sağlamak da mümkündür. Bunun için
    girilen sayıdan önce yazılan ek bilgilerle (ön ek - prefix) derleyicinin sayıyı değerlendireceği taban
    belirtilmektedir.

Farklı sayı tabanlarında verilen sayıları onluk tabana çevirmek için Convert.ToDecimal fonksiyonu
kullanılmaktadır. Aşağıda verilen program farklı tabanlarda yazılan 1011 değerinin onluk tabandaki

---
int onluk = 1011; // 10luk tabanda degerlendirilir.
int ikilik = 0B1011; // 2lik tabanda degerlendirilir.
int onaltilik = 0x1011; // 16lik tabanda degerlendirilir.
Console.WriteLine("Onlu tabanda 1011:\t" + Convert.ToDecimal(onluk));
Console.WriteLine("İkili tabanda 0B1011:\t" + Convert.ToDecimal(ikilik));
Console.WriteLine("Onaltılı tabanda 0x1011:\t" + Convert.ToDecimal(onaltilik));
Console.ReadKey();
Ondalıklı sayı veri tiplerine değer atarken ondalık ayıracı olarak . kullanılır.

---
### Karakter ve string Değerler

Karakter değerleri adı üzerinde alfabetik, nümerik veya alfanümerik karakter, karakter katarı/karakter dizisi
(dizge) olabilirler.
char veri tipi tek bir alfanümerik karakter değeri alabilir ve bu değer tek tırnak içinde '' verilmelidir. char veri
tipinin alabileceği değerler herhangi bir karakter, Unicode karakter kodu (\udddd formatında 4 haneli Unicode numarası), hexadecimal karakter kodu (\xdddd formatında) ve ASCII karakter kodları (0-255 arası)
olabilir. Aşağıda bu durum örneklenmiştir.



char karakter = 'C';
char unicodeKarakter = '\u0043';
char hexadecimalKarakter = '\x0043';
char asciiKarakter = (char)67;
Console.WriteLine($"Karakter: {karakter}, Unicode: {unicodeKarakter}, HexaDecimal:{hexadecimalKarakter}, ASCII: {asciiKarakter}");


string veri tipi ise birden fazla alfanümerik karakter alabilen veri tipidir. string değerler çift tırnak
karakterleri "" içinde verilmelidir. İlk çift tırnak, karakter dizgesinin verilmeye başlandığını; ikinci çift tırnak
ise karakter dizgesinin sonlandığını belirtir.

string dizge = "C#";

Bir dizge içinde çift tırnak karakteri olabilir. Bu durumda derleyici ikinci gördüğü çift tırnağı karakter
değerlerinin sonlandırıcısı olarak algılayacak ve sonraki kısım derleyiciye anlamsız geldiği için hata
verecektir.

string ifade = "C# ".NET" ortamında çalışan bir dildir."; // hatali

Bu gibi durumlarda derleyicinin hata vermemesi için (bu gibi karakterleri derleyiciden saklamak için) kaçış
karakteri adı verilen aşağıdaki tabloda verilen karakterler kullanılır. 

Ters bölü \ karakterinden sonra kullanılan bu kaçış karakterlerinin her birinin ayrı bir anlamı vardır. Dizgenin içinde çift tırnak varsa ve bunu derleyicinin metin içinde kullanılan bir çift tırnak olarak değerlendirmesi için çift tırnak karakterinden önce \ karakteri kullanılır. 

**Veya dizge içinde TAB boşluğu verilmek istenirse \t, dizgeye yeni bir satır eklemek için \n kaçış karakteri kullanılmalıdır.
**
```c#
string ifade = "C# \".NET\" ortamında çalışan bir dildir.";
```

### bool Değerler

bool değerler yalnızca true ve false olabilir. İlave olarak true veya false döndüren ifadeler de kullanılabilir. 1
veya 0 değerleri bool veri tipine değer olarak verilemez.

bool durum = true;
bool basarili = false;

bool kaydet = 10 > 5 && 3 >= -5;
bool karar = true & !false ^ false;


#### Bölüm Özeti
C# nesne yönelimli bir dildir ve kodun işleyişi sınıfların etrafında gerçekleşir. Sınıf içerisinde bir eylemi
yerine getiren ve ayrılmış kod bölümlerine metod denir. Bir sınıfta çeşitli sayıda metodlar tanımlanabilir. Bir
problemi çözümünde birden fazla sınıf tanımlamak gerekebilir. Alan uzayı, her bir sınıfta yapılan
isimlendirmelerin karışmaması ve sınıfları mantıksal bir çatı alında toplamak için kullanılan en geniş
kapsamlı yapıdır. Çalıştırılabilir kod kısımları birbirinden noktalı virgül ile ayrılır. Bir değer taşıyan veya
değer döndüren kod kısımlarına ifade denir. Bir işi gerçekleştiren kod kısımları ise deyim olarak adlandırılır.
Kaynak kodun programa dönüşmesi yani 1 ve 0’lardan oluşan makine diline dönüşmesi için derleyici adı
verilen programlar kullanılır. C# dili iki aşamalı derlenen bir dildir. İlk aşamada kaynak kod .NET
platformunda bir ara forma dönüştürülür ve ikinci aşamada bu ara form makine diline çevrilir.
C# çok amaçlı, modern, tip güvenli ne nesne yönelimli bir dildir. Geniş topluluk desteğine ve kullanım
oranına sahiptir.
C# dilinde özel bir işlevi olan kelimelere anahtar kelime denir ve bunlar başka amaçlarla kullanılamaz veya
programcı tarafından bir yapıya isim olarak verilemezler.
Kodun içinde kodun tanıdığı ve bildiği kelime, ifade ve operatörlerin dışında kodun ne yaptığını açıklamak,
bilgi vermek için yorum satırları kullanılır. Yorum satırlarını derleyici es geçer, değerlendirmez.
Programda bir işlevi yerine getiren bir yapıyı daha sonra tekrar kullanmak gerekebilir. Bu durumda bu
yapılara bir isim vermek gerekmektedir. Programcının yapılara verdiği bu isimlere tanımlayıcı/tanıtıcı denir.
Bir değişken veya fonksiyona isim vermek gelişigüzel yapılamaz, dilin tayin ettiği çeşitli isimlendirme
kuralları bulunmaktadır. Örneğin tanıtıcılar harfle başlamalı, boşluk ve özel karakter içermemeli, anahtar
kelime olmamalıdır.
Programda kullanılacak veriyi bilgisayarın belleğinde tutmak için değişkenler kullanılır. Adı üzerinde
değişkene atanan değer sonradan değiştirilebilir. Programda kullanılması icap eden fakat daha sonra
değiştirilmemesi gereken veri sabit olarak tanımlanmalıdır. Programda istenen her türlü veri ve değer
değişkenin veri tipine uyumlu olmalıdır.
