## Genel İfadeler

- Programlama dili nedir?

C# gibi programlama dilleri, bilgisayarın gerçekleştirmesini istediğiniz yönergeleri yazmanızı sağlar. Her programlama dilinin farklı bir söz dizimi vardır, ancak ilk programlama dilinizi öğrenip ikinci bir dil öğrenmeye çalıştığınızda, hepsinin birçok benzer fikri paylaştığını hemen fark edersiniz. Programlama dilinin işi, bir insanın yönergeleri insan tarafından okunabilir ve anlaşılır bir şekilde yazmasına izin vermektir. Programlama dilinde yazdığınız yönergelere "kaynak kod" veya yalnızca "kod" adı verilir.

Bu noktada yazılım geliştiricisi kodu güncelleştirebilir ve değiştirebilir, ancak bilgisayar kodu anlayamaz. İlk olarak kod, bilgisayarın anlayabileceği bir biçimde derlenmelidir .

- Derleme nedir?

Derleyici adlı özel bir program, kaynak kodunuzu bilgisayarın CPU'sunun yürütebileceği farklı bir biçime dönüştürür. Önceki ünitede yeşil Çalıştır düğmesini kullandığınızda, yazdığınız kod önce derlendi, sonra yürütüldü.

Kodumuzu neden derlememiz gerekiyor? Bilgisayar dillerinin çoğu başlangıçta şifreli gibi görünse de, bunlar bilgisayarın tercih ettiği dille karşılaştırıldığında insanlar tarafından daha kolay anlaşılabilir. Çünkü bilgisayarın tercih ettiği dil binlerce veya milyonlarca küçük anahtar açılarak veya kapatılarak ifade edilir. Derleyiciler insanlar tarafından okunabilen yönergeleri bilgisayar tarafından anlaşılabilen yönerge kümesine çevirerek iki dünya arasında köprü kurar.

- Söz dizimi nedir?

Bir bilgisayar dilinin söz dizimi, derleyicinin zorunlu tuttuğu anahtar sözcükleri, işleçleri (noktalı virgül veya parantez gibi özel klavye karakterleri) ve diğer dil bilgisi kurallarını içerir. Yazdığınız kod satırları bir düzine kadar söz dizimi kuralına uygundur ve bu satırlarda en az dört farklı işleç kullanılmıştır. Öğrenecek çok şey var, ancak neyse ki her kavram basittir. Vazgeçmeyin! ### Type (tür) systeme bakış

Aşağıdaki ana hat, C#'nin tür sistemine (type system) genel bir bakış sağlar.

- Değer türleri - Value Types

  - Basit türler - Simple Types
    - İmzalı tam tamsayı - Signed integral : `sbtye, short, int, long`
    - Unsigned integral- : `byte, ushort ,uint, ulong`
    - Unicode karakterleri : `char` UTF-16 kod birimini temsil eden
    - IEEE ikili kayan noktası - IEEE binary floating-point: `float, double`
    - Yüksek duyarlıklı ondalık kayan nokta - High-precision decimal floating-point: `decimal`
    - Boolean: `bool` Boole değerlerini true ve false sonucu verir.
  - Enum türleri
    - formunun kullanıcı tanımlı türleri `enum E {...} .` Tür enum , adlandırılmış sabitlere sahip ayrı bir tür olur. Her enum türün, sekiz tamsayı türünden biriolması gereken bir temel türü vardır. Bir türün değer `enum` kümesi, temelalınan türün değer kümesiyle aynıdır.
  - Yapı türleri - Struct Type
    - Formun kullanıcı tanımlı türleri `struct S {...}`
  - Boş değer atanabilen değer türleri - Nullable value types
    Değere sahip diğer tüm değer türlerinin uzantıları `null`
  - Tuple değer türleri
    Formun kullanıcı tanımlı türleri `(T1, T2, ...)`

- Başvuru türleri
  - Sınıf türleri - class type
    - Diğer tüm türlerin nihai temel sınıfı: `object`
    - Unicode dizeleri: `string `UTF-16 kod birimleri dizisini temsil eden
    - Formun kullanıcı tanımlı türleri `class C {...}`
  - Arabirim türleri - Interface types
    - Formun kullanıcı tanımlı türleri `interface I {...}`
  - Dizi türleri - Array types
    - Tek boyutlu, çok boyutlu ve etiketli. Örneğin: int[] , int[,] ve int[][]
  - Temsilci türleri
    - Formun kullanıcı tanımlı türleri delegate `int D(...)`

C# programları, yeni türler oluşturmak için tür bildirimlerini kullanır. Tür bildirimi, yeni türün adını ve üyelerini belirtir. C# kategorisinin altı türü kullanıcı tarafından tanımlanabilir: sınıf türleri, yapı türleri, arabirim türleri, enum türleri, temsilci türleri ve tuple değer türleri. ## Değişkenler

Sabit değer sözcük anlamıyla sabit kodlanmış değerdir. Öte yandan çoğu uygulama önceden pek fikir sahibi olmadığımız değerlerle çalışmamızı gerektirir. Başka bir deyişle kullanıcılardan, dosyalardan veya ağ üzerinden gelen verilerle çalışmamız gerekir.

Kodumuzun dışından gelen verilerle çalışmamız gerektiğinde bir değişken bildiririz.

Değişken nedir?

Değişken, kullanım ömrü boyunca değerini değiştirebilen bir veri öğesidir. Kodunuzda daha sonra kullanmayı planladığınız değerleri geçici olarak depolamak için değişkenleri kullanabilirsiniz. Değişken, bilgisayar bellek adresine atayabildiğimiz kolay bir etikettir. Bir değeri geçici olarak bu bellek adresinde depolamak istediğimizde veya bellek adresinde depolanan değeri almak istediğimizde yalnızca oluşturduğumuz değişken adını kullanacağız.

Değişken bildirme

Yeni değişken oluşturmak için önce değişkenin veri türünü bildirmeli ve ardından buna bir ad vermeliyiz.


`````python
string firstName;
`````

Buradaki örnekte string türünde firstName adlı yeni bir değişken oluşturuyoruz. Bundan sonra bu değişken yalnızca dize değerlerini tutabilir.

Değişkenleri adlandırmak için birkaç C# söz dizimi kuralına uygun olduğu sürece herhangi bir ad seçebilirsiniz.

Değişken adı kuralları
Bir yazılım geliştiricisi ünlü bir şekilde şunları söyledi: "Yazılım geliştirmenin en zor kısmı, öğeleri adlandırmaktır." Bir değişkenin adının belirli söz dizimi kurallarına uyması gerekmez, aynı zamanda kodu daha insan tarafından okunabilir ve anlaşılır hale getirmek için de kullanılmalıdır. Bir kod satırından istenen ne çok şey var!

Aşağıda değişken adları hakkında birkaç önemli noktayı bulabilirsiniz:

- Değişken adları, alfasayısal karakterler ve alt çizgi karakterini içerebilir. Karma sembolü # (sayı sembolü veya sterlin sembolü olarak da bilinir) ya da dolar sembolü $ gibi özel karakterlere izin verilmez.
- Değişken adları, rakamla değil, harf veya alt çizgi ile başlamalıdır. Geliştiriciler, alt çizgiyi özel bir amaçla kullanır, bu nedenle şimdilik bunu kullanmamaya çalışın.
- Değişken adları bir C# anahtar sözcüğü olmamalıdır. Örneğin, aşağıdaki değişken bildirimlerini kullanamazsınız: decimal decimal; veya string string;.
- Değişken adları büyük/küçük harfe duyarlıdır. Bu, string Value; ile string value; öğelerinin iki farklı değişken olduğu anlamına gelir.
- Değişken adlarının orta harfi büyük olmalı; başka bir deyişle, ilk sözcüğün başında küçük harf ve sonraki her sözcüğün başında büyük harf kullanılmalıdır. Örneğin: string thisIsCamelCase;.
- Değişken adları, açıklayıcı ve uygulamanızda anlamlı olmalıdır. Değişkeniniz için, tutacağı verilerin türünü gösteren bir ad seçin.
- Değişken adları, birleştirilmiş bir veya daha fazla tam sözcükten oluşmalıdır. Değişkenin adı (ve dolayısıyla amacı) kodunuzu okuyan diğer kişiler için belirsiz olabileceğinden, kısaltmaları kullanmayın.
- Değişken adları, değişkenin veri türünü içermemelidir. string strValue; gibi bir stil kullanmanız yönünde öneriler görebilirsiniz. Bu öneri artık geçerli değil.

Birinin adını temsil eden verileri depolamak için bu değişkeni kullanmak istediğiniz varsayılarak, örnek string firstName; bu kuralların ve kuralların tümünü izler.

Örnekler:


`````python
string yaziKarakteri;
yaziKarakteri = "Buraya istediğimiz metni yazabiliriz";
Console.WriteLine(yaziKarakteri);


string adi, soyadi;
//string soyadi
int age;

adi = "Cemil";
soyadi = "Kerim";
age = 35;

Console.Write(adi);
Console.WriteLine(soyadi);
Console.Write(age);
`````

## Örtülü olarak belirtilmiş yerel değişkenler

Kodunuzu yazarken C# derleyicisi birçok kolaylık sağlayabilir. Başlatılan değere göre değişkeninizin veri türünü çıkarsayabilir. Bu ünitede, örtük olarak yazılan yerel değişkenler olarak adlandırılan bu özellik hakkında bilgi edineceksiniz.

Örtülü olarak belirtilmiş yerel değişkenler nedir?

Örtülü olarak belirtilmiş yerel değişken, C# derleyicisine türü çıkarsamasını bildiren var anahtar sözcüğü kullanılarak oluşturulur. Tür çıkarıldıktan sonra, değişkeni bildirmek için gerçek veri türünü kullanmış olmanızla aynıdır.

Aşağıdaki örnekte string anahtar sözcüğü yerine var anahtar sözcüğünü kullanarak bir değişken bildireceğiz.


`````python

var message = "Hello world!"
`````

değişkeni message hemen dize değerine "Hello World!"ayarlandığından, C# derleyicisi amacı anlar ve iletinin her örneğini türünün stringbir örneği olarak değerlendirir.

Gerçekten message değişkeni dize olacak şekilde yazılır ve hiçbir zaman değiştirilemez. Burada message değişkenini 10.0m sabit ondalık değerine ayarlamaya çalışacağız.

Yer Tutucu Yöntemiyle Çıktılama

Konsol arayüzünde değişken(ler) ve metin aynı satırda yer tutucu (string placeholder) yöntemiyle ekrana
çıktılanabilir. Yer tutucu bir değişken veya bir değerin yerini tutar. Çift tırnakların arasında süslü parantezler
verilir ve içinde ekrana çıktılanacak değişkenin numarası yazılır. Çift tırnak bittikten sonra virgül verilir ve
çıktılanacak değişken(ler) yazılır. Verilen ilk değişkenin numarası sıfırdır, sonrakiler bir artarak devam
ederler. Temel sözdizimi aşağıda verilmiştir

```c#
"{0}", degisken0/deger0
"{0}{1}",degisken0/deger0, degisken1/deger1
```

Aşağıdaki programda yer tutucu yöntemi kullanılarak iki değişken aynı satırda çıktılanmıştır. Verilen kodda 0 numaralı değişken bolum, 1 numaralı değişken ders değişkenidir.


`````python
string bolum = "Bilgisayar Programciligi";
string ders = "c# programlama";
Console.Write("{0} - {1}", bolum, ders);
`````

Süslü parantezler içinde yer tutucu numaralarının düzenli sırada verilmesi zorunlu değildir. Yukarıdaki programda yer tutucu numaraları 0 ve 1 olarak verilmiştir. Fakat kod içinde yer tutucu numaraları aşağıdaki satırda olduğu gibi 1 ve 0 şeklinde değiştirildiğinde program çıktısı da değişecektir. İlk değişkenin numarasının 0, sonrakinin 1 ve böylece artarak gittiği hatırlandığında 0 numaralı değişken bolum, 1 numaralı değişken ders değişkenidir. Bu durumda önce 1 numaralı değişken olan ders değişkeni ve daha sonra 0 numaralı değişken olan bolum değişkeni çıktılanacaktır. Bu durumda programın çıktısı aşağıdaki gibi olacaktır.


`````python
string bolum = "Bilgisayar Programciligi";
string ders = "C# programlama";
Console.Write("{1} - {0} ", bolum, ders);
`````

`````python
string bolum = "Bilgisayar Programciligi";
string ders = "C# programlama";
string fak = "auzef";
Console.WriteLine("{1} - {0} - {2} ", fak, ders, bolum);
Console.WriteLine("{1} - {0} - {2} ", ders, bolum, fak);
Console.WriteLine("{2} - {1} - {0} ", bolum, ders, fak);
`````

### Console.Read() :

Kullanıcıdan tek bir karakter almak için kullanılır. Girilen karakterlerin kendisi değil **ASCI kodu int** veri tipinde tutulur.

```
Console.WriteLine("Bir karakter giriniz: ");
int ch = Console.Read();
Console.Write("Girdiğiniz karakterin ASCI kodu : {0}", ch);

```

Girilen karakterin ASCII kodunu değil de karakterin kendisini ekrana çıktılamak için int veri tipinde tutulan kullanıcı girdisi `Convert.ToChar()` metoduyla char veri tipine çevrilmelidir

```
Console.WriteLine("Bir karakter giriniz: ");
int ch = Convert.ToChar(Console.Read());
Console.Write("Girdiğiniz karakterin ASCI kodu : {0}", ch);
```

---

### Console.ReadKey() :

Kullanıcıdan tek bir tuş verisi almak için kullanılır. Genellikle programın yapacağı işler bittikten sonra programı sonlandırmak için kullanıcıdan bir tuşa basması istenir. Bu amaçla Console.ReadKey() metodu kullanılır.

```

ConsoleKeyInfo tus = Console.ReadKey();

Console.ReadKey();

```

Console.WriteLine("Bir metin giriniz : ");
string metin = Console.ReadLine();

#### Erişim Belirleyiciler : 

Bazen program içindeki tip, metod, değişken, struct gibi yapıların son kullanıcı tarafından veya sınıfın bazı
metodları tarafından veya diğer sınıflar tarafından erişilememesini istediğimiz durumlar olabilir. Örneğin
bazı değişkenler sınıf içi yapıyı tanzim etmek için kullanılırlar bu nedenle dışarıdan bu değişkenlerin
değerlerinin görüntülenmesi anlamsız olacaktır. Kimi zaman bazı metodlar gizli bilgi taşırlar ve bu nedenle
sadece izinli metodların bu bilgiye erişiminin olmasını isteriz, diğer metod veya yapıların bu metoda
erişmesini istemeyiz. İşte bu amaçla bazı programlama yapılarına erişimi sınırlamak için erişim
belirleyiciler (access modifiers) adı verilen 4 farklı yapı 6 farklı seviye kullanılmaktadır. Aşağıda listelenen
erişim belirleyiciler tip, değişken, metod ve struct dil yapılarının sınıf içinde, sınıf dışında, proje içinde veya
dışında görünürlüğünü etkilemektedir.

**public**:

public anahtar kelimesi ile tanımlanan bir dil bileşeni her yerden görünür ve erişilebilirdir. Tanımlandığı
sınıftan, diğer sınıflardan, farklı derlemedeki sınıflar tarafından erişilebilir ve kullanılabilir.

**private**:

private anahtar kelimesi ile tanımlanan bir dil bileşeni yalnızca tanımlandığı sınıf veya struct tarafından
erişilebilir. private, değişkenlerin varsayılan erişim belirleyicisidir.

**protected**:

protected anahtar kelimesi ile tanımlanan sınıf, metod veya sınıf değişkenleri yalnızca tanımlandığı sınıf
veya bu sınıftan türetilmiş sınıflardan veya farklı derlemede yer alan türetilmiş sınıflardan erişilebilir.

**internal**:

internal anahtar kelimesi ile tanımlanan bir dil bileşeni hangi kod bileşenleri ile birlikte derlendiyse onlar
tarafından erişilebilir. internal, sınıfların varsayılan erişim belirleyicisidir.

**protected internal**:

protected internal anahtar kelimesi ile tanımlanan bir dil bileşeni aynı derlemede yer alan sınıf veya
türetilmiş sınıflar tarafından veya başka bir derlemede yer alan türetilmiş sınıf tarafından erişilebilir.

**private protected**:

private protected anahtar kelimesi ile tanımlanan bir dil bileşeni derlendiği sınıftan veya aynı derlemede
türetilmiş diğer sınıflar tarafından erişilebilir

C# dilinde veri tiplerinin alabileceği değerler aşağıda kategoriler halinde açıklanmıştır.

    **Nümerik Değerler:**

    * Nümerik veri tiplerine değer girilirken rakamlar kullanılır. 
    * Eğer sayı onaltılık tabanda ise ilave olarak A-F arası harfler, 
    * sayının ondalık ayıracı olarak nokta . operatörü, 
    * bilimsel gösterim (notasyon) için E harfi,
    * sayının okunurluğunu artırmak için istenirse binlik ayıracı olarak alt çizgi _ karakteri kullanılmaktadır.

    Bunların haricinde boşluk, virgül karakterleri nümerik değerlerde kullanılmaz. 
    long, float ve decimal veri tipi
    tercih edilmişse bu veri tiplerine girilen değerin sonuna sırasıyla (küçük veya büyük) L, F ve M harfleri
    eklenmelidir. Sayının işareti için – ve + operatörleri kullanılır.

    Günlük hayatta hesaplamalarımızı genellikle onluk sistem üzerinden yaparız. C# dilinde de tam sayı veri
    tiplerine değer atanırken girilen değerler varsayılan olarak onluk tabanda değerlendirilmektedir. Bununla
    birlikte değişkene atanan sayının farklı tabanlarda değerlendirilmesini sağlamak da mümkündür. Bunun için
    girilen sayıdan önce yazılan ek bilgilerle (ön ek - prefix) derleyicinin sayıyı değerlendireceği taban
    belirtilmektedir.

Tablo 3: Sayı tabanı belirten ön ekler

Taban Ön Ek Kullanılabilecek Rakam ve Harfler Örnek
Onluk Yok. Varsayılan onluk tabandır. 0-9 int onluk = 38910;
İkilik 0b veya 0B 0-1 int ikilik = 0B1100;
Onaltılık 0x veya 0X 0-9, A-F int onaltilik = 0x98A5F0;


| Taban  	|Ön Ek   	    |Kullanılabilecek Rakam ve Harfler   	|Örnek	                   |   	
|---	    |---	        |---	                                |---	                   |
|Onluk   	|  Yok 	        |  0-9 	                                |int onluk = 38910;   	   |  
|İkilik	    |0b veya 0B    	|  0-1	                                |int ikilik = 0B1100;      | 
|Onaltılık  |0x veya 0X    	| 0-9,A-F	                            |int onaltilik = 0x98A5F0; | 


Farklı sayı tabanlarında verilen sayıları onluk tabana çevirmek için Convert.ToDecimal fonksiyonu
kullanılmaktadır. Aşağıda verilen program farklı tabanlarda yazılan 1011 değerinin onluk tabandaki
karşılığını ekrana çıktılamaktadır.


---

`````python
int onluk = 1011; // 10luk tabanda degerlendirilir.
int ikilik = 0B1011; // 2lik tabanda degerlendirilir.
int onaltilik = 0x1011; // 16lik tabanda degerlendirilir.
Console.WriteLine("Onlu tabanda 1011:\t" + Convert.ToDecimal(onluk));
Console.WriteLine("İkili tabanda 0B1011:\t" + Convert.ToDecimal(ikilik));
Console.WriteLine("Onaltılı tabanda 0x1011:\t" + Convert.ToDecimal(onaltilik));
Console.ReadKey();
`````

Ondalıklı sayı veri tiplerine değer atarken ondalık ayıracı olarak . kullanılır.

```c#
float ondalikli = 3.14F;
decimal taksitliFiyat = 35000.5M;
```

Sayının negatif veya pozitif olduğunu bildirmek için **– ve + **işaret operatörleri kullanılmaktadır.

```c#
short pozitif = 9;
sbyte negatif = -15;
decimal bakiye = +1.5E4M;
```

Günlük hayatta çok basamaklı sayıları yazarken okuma kolaylığı olması için her üç basamakta bir binlik
ayıracı olan nokta . kullanırız. Nokta . operatörü C# dilinde binlik değil ondalık ayıracı olarak
kullanılmaktadır. Nümerik değerlerin okunurluğunu artırmak için binlik ayıracı olarak alt çizgi _
kullanılabilir.

```c#
long buyukSayi = 11_230_000_000L;
Console.WriteLine("Binlik ayıracıyla yazılmış sayımız: " + buyukSayi);
```

Çok basamaklı sayıların bilimsel notasyonda 10’un kuvveti olarak gösterilmesi için e veya E harfi
kullanılmaktadır. Bilimsel notasyon tam sayı veri tipleriyle kullanılamaz. E harfinden sonra sayının kuvveti
verilmelidir.

```c#
double bilimselNotasyon = 11.23E9;
Console.WriteLine("Bilimsel notasyonda yazılmış sayımız: " + bilimselNotasyon);
```

Nümerik veri tiplerinde tanımlanan değişkenlere veri tipinin değer aralığının dışında değer girilmesi halinde

#### Karakter ve string Değerler

Karakter değerleri adı üzerinde alfabetik, nümerik veya alfanümerik karakter, karakter katarı/karakter dizisi
(dizge) olabilirler.
char veri tipi tek bir alfanümerik karakter değeri alabilir ve bu değer tek tırnak içinde '' verilmelidir. char veri
tipinin alabileceği değerler herhangi bir karakter, Unicode karakter kodu (\udddd formatında 4 haneli Unicode numarası), hexadecimal karakter kodu (\xdddd formatında) ve ASCII karakter kodları (0-255 arası)
olabilir. Aşağıda bu durum örneklenmiştir.




`````python
char karakter = 'C';
char unicodeKarakter = '\u0043';
char hexadecimalKarakter = '\x0043';
char asciiKarakter = (char)67;
Console.WriteLine($"Karakter: {karakter}, Unicode: {unicodeKarakter}, HexaDecimal:{hexadecimalKarakter}, ASCII: {asciiKarakter}");

`````

string veri tipi ise birden fazla alfanümerik karakter alabilen veri tipidir. string değerler çift tırnak
karakterleri "" içinde verilmelidir. İlk çift tırnak, karakter dizgesinin verilmeye başlandığını; ikinci çift tırnak
ise karakter dizgesinin sonlandığını belirtir.

string dizge = "C#";

Bir dizge içinde çift tırnak karakteri olabilir. Bu durumda derleyici ikinci gördüğü çift tırnağı karakter
değerlerinin sonlandırıcısı olarak algılayacak ve sonraki kısım derleyiciye anlamsız geldiği için hata
verecektir.

string ifade = "C# ".NET" ortamında çalışan bir dildir."; // hatali

Bu gibi durumlarda derleyicinin hata vermemesi için (bu gibi karakterleri derleyiciden saklamak için) kaçış
karakteri adı verilen aşağıdaki tabloda verilen karakterler kullanılır. 

Ters bölü \ karakterinden sonra kullanılan bu kaçış karakterlerinin her birinin ayrı bir anlamı vardır. Dizgenin içinde çift tırnak varsa ve bunu derleyicinin metin içinde kullanılan bir çift tırnak olarak değerlendirmesi için çift tırnak karakterinden önce \ karakteri kullanılır. 

**Veya dizge içinde TAB boşluğu verilmek istenirse \t, dizgeye yeni bir satır eklemek için \n kaçış karakteri kullanılmalıdır.
**
```c#
string ifade = "C# \".NET\" ortamında çalışan bir dildir.";
```
|  İşaret    |   Sembol |
|:---:|:---:|
|   \'     |   Tek tırnak |
|   \"     |   Çift tırnak|
|   \\     |   Ters bölü|
|   \0     |   Null|
|   \a     |   Uyarı|
|   \b     |   Backspace tuşundan kaçış karakteri|
|   \f     |   Form besleme|
|   \n     |   Yeni satır|
|   \r     |   Satır başı|
|   \t     |   Yatay sekme (TAB)|
|   \v     |   Dikey sekme (TAB)|
|   \u     |   UTF-16 Unicode kaçış karakteri|
|   \U     |   UTF-32 Unicode kaçış karakteri|
|   \x     |   Değişken uzunluk haricinde "\u"ya benzer Unicode çıkış karakteri|



Aşağıda dizge içinde çift tırnak olan ilk satır hata verirken, kaçış karakteriyle çift tırnağı perdeleyen ikinci
satır hata vermeyecektir.

```c#
string ifade = "C# ".NET" ortamında çalışan bir dildir."; // hatali
string ifadeKacisli = "C# \".NET\" ortamında çalışan bir dildir."; // duzgun

```
Metin içinde özel anlamı olan kaçış karakterlerini çıktılamak için her seferinde ters bölü karakteri \ eklemek
zor olabilir. Bunun yerine karakter dizgelerine değer atarken çift tırnak " karakterinden önce @ operatörü
kullanılabilir. Bu sayede metin olduğu gibi çıktılanacak, derleyici metin içindeki karakterlere özel bir anlam
yüklemeyecektir. Aşağıda \ karakteri içeren dizge ilkinde kaçış karakteri ile ikincisinde ise @ operatörü ile
değişkenlere atanmıştır.

```c#
string sistemKlasoru = "C:\\Windows\\System32";
string sistemKlasoruVerbatim = @"C:\Windows\System32";
```

@ operatörü ile tanımlanan dizgenin içinde çift tırnak " karakteri yer alamaz. Derleyicinin hata vermemesi
için çift tırnak " karakterinden önce tekrar bir çift tırnak " karakteri verilmelidir.

```c#
string ifadeHatali = @"C# ".NET" ortaminda calisan bir dildir."; // hatali
string ifadeDuzgun = @"C# "".NET"" ortaminda calisan bir dildir."; // duzgun
```

Çok satırlı dizge tanımlamak için her bir satırın sonunda \n kaçış karakteri verilmeli, ikinci satırın dizgesi +
operatörü ile ilkine eklemlenmelidir.

```c#
string cokSatirliDizge = "Ilk satir\n" +
"Ikinci satir";
@ operatörü ile çok satırlı karakter dizgeleri çok daha kolay bir şekilde tanımlanabilir.
string cokSatirliDizgeVerbatim = @"Ilk satir
Ikinci satir";
Console.WriteLine(cokSatirliDizgeVerbatim);
```

#### bool Değerler

bool değerler yalnızca true ve false olabilir. İlave olarak true veya false döndüren ifadeler de kullanılabilir. 1
veya 0 değerleri bool veri tipine değer olarak verilemez.

bool durum = true;
bool basarili = false;

bool kaydet = 10 > 5 && 3 >= -5;
bool karar = true & !false ^ false;



#### Bölüm Özeti
C# nesne yönelimli bir dildir ve kodun işleyişi sınıfların etrafında gerçekleşir. Sınıf içerisinde bir eylemi
yerine getiren ve ayrılmış kod bölümlerine metod denir. Bir sınıfta çeşitli sayıda metodlar tanımlanabilir. Bir
problemi çözümünde birden fazla sınıf tanımlamak gerekebilir. Alan uzayı, her bir sınıfta yapılan
isimlendirmelerin karışmaması ve sınıfları mantıksal bir çatı alında toplamak için kullanılan en geniş
kapsamlı yapıdır. Çalıştırılabilir kod kısımları birbirinden noktalı virgül ile ayrılır. Bir değer taşıyan veya
değer döndüren kod kısımlarına ifade denir. Bir işi gerçekleştiren kod kısımları ise deyim olarak adlandırılır.
Kaynak kodun programa dönüşmesi yani 1 ve 0’lardan oluşan makine diline dönüşmesi için derleyici adı
verilen programlar kullanılır. C# dili iki aşamalı derlenen bir dildir. İlk aşamada kaynak kod .NET
platformunda bir ara forma dönüştürülür ve ikinci aşamada bu ara form makine diline çevrilir.
C# çok amaçlı, modern, tip güvenli ne nesne yönelimli bir dildir. Geniş topluluk desteğine ve kullanım
oranına sahiptir.
C# dilinde özel bir işlevi olan kelimelere anahtar kelime denir ve bunlar başka amaçlarla kullanılamaz veya
programcı tarafından bir yapıya isim olarak verilemezler.
Kodun içinde kodun tanıdığı ve bildiği kelime, ifade ve operatörlerin dışında kodun ne yaptığını açıklamak,
bilgi vermek için yorum satırları kullanılır. Yorum satırlarını derleyici es geçer, değerlendirmez.
Programda bir işlevi yerine getiren bir yapıyı daha sonra tekrar kullanmak gerekebilir. Bu durumda bu
yapılara bir isim vermek gerekmektedir. Programcının yapılara verdiği bu isimlere tanımlayıcı/tanıtıcı denir.
Bir değişken veya fonksiyona isim vermek gelişigüzel yapılamaz, dilin tayin ettiği çeşitli isimlendirme
kuralları bulunmaktadır. Örneğin tanıtıcılar harfle başlamalı, boşluk ve özel karakter içermemeli, anahtar
kelime olmamalıdır.
Programda kullanılacak veriyi bilgisayarın belleğinde tutmak için değişkenler kullanılır. Adı üzerinde
değişkene atanan değer sonradan değiştirilebilir. Programda kullanılması icap eden fakat daha sonra
değiştirilmemesi gereken veri sabit olarak tanımlanmalıdır. Programda istenen her türlü veri ve değer
değişkenin veri tipine uyumlu olmalıdır.
