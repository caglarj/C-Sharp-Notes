## C# Sabitleri (Const)

Değişkenler bölümünde açıklandığı gibi, C# içindeki bir değişken belirtilen bir veri türü olmalıdır:



## C# Type Casting - Tür Dönüşümleri :

bir veri türünün değerini başka bir türe atanması olarak tanımlayabiliriz. 

* Implicit Casting  (otomatik) - daha küçük bir türü dönüştürme daha büyük bir tür boyutuna
    char -> int -> long -> float -> double

```c#
int myInt = 9;
double myDouble = myInt;       // Automatic casting: int to double

Console.WriteLine(myInt);      // Outputs 9
Console.WriteLine(myDouble);   // Outputs 9
```


* Explicit Casting (manuel) - daha büyük bir türü dönüştürme daha küçük boyutlu bir türe
    double -> float -> long -> int -> char

```c#
double myDouble = 9.78;
int myInt = (int) myDouble;    // Manual casting: double to int

Console.WriteLine(myDouble);   // Outputs 9.78
Console.WriteLine(myInt);      // Outputs 9
```


Örneğin tam sayı veri tipinde tanımlanan bir değişkene karakter verisi girilemeyecektir. Böylece sonuçların tutarlı olması sağlanırken istenmeyen sonuçların üretilmesi daha baştan önlenmektedir.

C# dilinde veri tipleri Değer veri tipi (Value Type) ve Referans veri tipi (Reference Type) olmak üzere iki kategoride sınıflandırılmaktadır (Microsoft, 2023a). Değer ve referans veri tipleri arasında bilinmesi gereken en temel fark; bilgisayar belleğinde değer veri tipinde tanımlanan bir değişkenin değerinin kendisi tutulurken, referans veri tipinde tanımlanan bir değişkenin bellekte değeri değil adresi tutulur. Değer veri tiplerine erişim hızlıdır, referans veri tipleri değişkenin değerini değil adresini sakladığından biraz daha yavaştır.

NET üzerinde sadece C# dili değil Visual Basic, F# gibi dillerde de kodlama yapılabilmektedir. .NET platformunda bu dillerin her biri için ayrı ayrı veri tipi tanımlanmamış bunun yerine Ortak Tip Sistemi (Common Type System-CTS) altında tip yapısı ve veri tipleri tanımlanmıştır. C# ve diğer .NET tabanlı diller Ortak Tip Sistemi altında tanımlanan veri tiplerini kullanırlar.


Değişkenler bölümünde açıklandığı gibi, C# içindeki bir değişken belirtilen bir veri türü olmalıdır:
